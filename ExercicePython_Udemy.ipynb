# **Déterminer le type d'une variable (texte, nombre, booléen...).**
Nous avons les variables ci-dessous.

Il est possible d'utiliser sur Python les f-strings pour afficher le contenu d'une variable directement dans une chaîne de caractères. Pour cela, il suffit de mettre dans une chaîne de caractères la variable que l'on souhaite afficher, entourée d'accolades.

# **Arrondir un nombre facilement.**
Nous avons une variable qui contient le nombre : 14.6749813.

Arrondir ce nombre avec seulement 2 décimales en utilisant un f-string, le tout dans un print()

Résultat attendu :

Nombre avec deux décimales : 14.67

```python
nombre = 14.6749813
print(f'Nombre avec deux décimales : {nombre:.2f}')
```

# **Inverser l'ordre des lettres d'un texte.**
Nous avons le texte ci-dessous :

"Ceci est un cours pour apprendre Python par la pratique !"

Inverser l'ordre des lettres de ce texte.

Afficher le résultat dans la console.

Résultat attendu :

! euqitarp al rap nohtyP erdnerppa ruop sruoc nu tse iceC

```python
# On définit une chaîne de caractères "texte_a_inverser" contenant un texte
texte_a_inverser = 'Ceci est un cours pour apprendre Python par la pratique !'

# On utilise une technique de slicing pour inverser la chaîne de caractères
texte_final = texte_a_inverser[::-1]

# On affiche la chaîne de caractères inversée
print(texte_final)
```
